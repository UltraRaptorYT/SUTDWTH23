```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("FoodKeeper-Data.xls", sheet_name="Product")
```

```python
df.isna().sum()
```

```python
columns = list(df.columns)
minCols = [x for x in columns if "Min" in x]
maxCols = [x for x in columns if "Max" in x]
```

```python
def replace_columns():
    for x_min in minCols:
        df["Pantry_Min"].fillna(df[x_min], inplace= True)
        df["Pantry_Min"] = df["Pantry_Min"].apply(str) + df["Pantry_Metric"]
    
    for x_max in maxCols:
        df["Pantry_Max"].fillna(df[x_max], inplace= True)
        df["Pantry_Min"] += df["Pantry_Min"].apply(str) + df["Pantry_Metric"]
```

```python
replace_columns()
```
